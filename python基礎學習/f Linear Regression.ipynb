###### TRAINING and TESTING data

```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.linspace(0,5,100)
y = 1.2*x + 0.8 + 0.5*np.random.randn(100)
plt.scatter(x,y)
```

# 標準函數訓練及測試

 - ###### 訓練資料TRAINING data 喂電腦資料用以學函數
 - ###### 測試資料TESTING data
 
 ###### &rarr; from sklearn.model_selection import train_test_split

```python
from sklearn.model_selection import train_test_split
```

###### ( x, y )資料分 80%為 TRAIN data ,且 20%為 TEST data

###### TRAININD data 和 TESTING data 分割
###### 輸入訓練資料 , 輸入測試資料 , 輸出訓練資料 , 輸出測試資料 = train_test_split( x, y, test_size = 分割給測試的比例, random_state = 隨機數種子)
隨機數種子其實就是該組隨機數的編號, 在需要重複試驗的時候，保證得到一組一樣的隨機數

```python
x_train, x_test, y_train, y_test = train_test_split(x,y,test_size = 0.2,random_state = 19)
```

```python
x_train = x_train.reshape(80,1)
x_test.shape = (20,1)
```

```python
from sklearn.linear_model import LinearRegression
```

```python
regr = LinearRegression()
```

```python
regr.fit(x_train,y_train)
```

```python
plt.scatter(x_train,y_train)
plt.plot(x_train,regr.predict(x_train),'r')
```

```python
plt.scatter(x_test,y_test)
plt.plot(x_test,regr.predict(x_test),'r')
```

###### predict Boston price

```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
```

讀Boston房價數據

```python
from sklearn.datasets import load_boston
```

features_name 數據參考的變數

```python
boston = load_boston()
```

```python
boston.feature_names
```

```python
X = boston.data
Y = boston.target
```

```python
len(X)
```

```python
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size= 0.3, random_state= 19)
```

```python
regr = LinearRegression()
```

```python
regr.fit(X_train,Y_train)
```

```python
Y_predict = regr.predict(X_test)
```

```python
plt.scatter(Y_test,Y_predict)
plt.plot([0,50],[0,50],'r')
plt.xlabel('True Price')
plt.ylabel('Predict Price')
```
