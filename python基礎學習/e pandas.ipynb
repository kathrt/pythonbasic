###### BASIC DATA ANALYSIS

```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
df = pd.read_csv('pythonpractice.csv')
```

```python
df.head()
```

```python
df.describe() #統計數據
```

```python
df.corr() #相關係數
```

```python
df.香農指數.corr(df.辛普森指數)
```

```python
df['加總'] = df[['香農指數','辛普森指數','百分比']].sum(1) #橫排為1 直列為0
```

```python
df.加總.head()
```

```python
df['加權'] = df.香農指數*1.4 + df.辛普森指數*1.6
```

```python
df.加權.head()
```

```python
df.sort_values(by = '加總',ascending= False).head(20) # 用加總排序,False為由大到小
```

```python
df.sort_values(by = ['加總','加權'],ascending= False).head() # 先排序'加總'再排序'加權'的分數但是要用[]隔開ascending
```

###### GROUPBY

```python
df = pd.read_csv('http://bit.ly/uforeports')
```

```python
df.head()
```

```python
df_state = df.groupby('State').count() # 以州計算看到UFO的次數
```

```python
df_state.head()
```

```python
df_state.sort_values(by = 'Time',ascending= False).head() #看排序但不改變不儲存為新的df_state
```

```python
df_state.head()
```

```python
df_state.sort_values(by = 'Time',ascending=False,inplace=True) #inplace
```

```python
df_state.head()
```

###### 畫長條圖

###### 股票分析

```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import pandas_datareader.data as web
```

```python
df = web.DataReader('AAPL','yahoo',start = '2012-9-1',end = '2017-8-31')
```

```python
df
```

```python
P = df['Adj Close']
```

```python
P.head()
```

```python
P.plot()
```

```python
P.diff() #報酬率
```

```python
r = P.diff()/P #報酬率與前一期百分比
```

```python
r.plot()
```

```python
r[-100:].plot()
```

```python
P.rolling(window = 20).mean()
```

```python
P.rolling(window = 20).mean().plot()
```

```python
P.plot()
P.rolling(window = 20).mean().plot()
P.rolling(window = 60).mean().plot()
```
