```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
```

```python
x0 = 0
y0 = 1
xp = 1
n  = 5        # 改n = 5, 10, 20
step_h = 0.2   # 只要改step_h = 0.2, 0.1, 0.05
```

```python
x = np.linspace(x0, xp, n)
y = np.zeros([n])
```

```python
y[0] = y0
for k in range(1, n):
    y[k] = y[k-1]+step_h*(np.e**(x[k-1]))
```

```python
for k in range(n):
    print(x[k], y[k])
```

```python
s = np.linspace(0, 1, 100)
plt.plot(x, y, 'o')
plt.plot(s, np.e**s)
plt.xlabel('value of x')
plt.ylabel('value of y')
plt.title('Approximate Solution with Euler’s Method h = 0.2')
plt.grid()
plt.show()
```
